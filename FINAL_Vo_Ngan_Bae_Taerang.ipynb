```python
import turtle 
import time 

scr=turtle.getscreen()
scr.title("American Flag")
scr.bgcolor("#FFFDD0")

t=turtle.Turtle() #initializing turtle function
t.speed(70) #setting the pen speed to 70
t.penup()

########## American Flag ##########
## stripes formatting
def draw_rectangle(x,y,height,width,color):
    t.goto(x,y)
    t.pendown()
    t.color(color)
    t.begin_fill()
    
    t.forward(width)
    t.right(90)
    t.forward(height)
    t.right(90)
    t.forward(width)
    t.right(90)
    t.forward(height)
    t.right(90)
    
    t.end_fill()
    t.penup()

## stripes positioning
def draw_stripes():
    x = -152
    y = 108
    height = 200 
    width = height*1.9 #following the G-spec dimensions

## red and white stripes (total 13 stripes in flag)
    stripe_ht = height/13
    stripe_wdt = width

    # drawing 6 red and 7 white strips
    for stripe in range(0,6):
        draw_rectangle(x,y, stripe_ht, stripe_wdt, '#B22234')
        # decreasing value of y by stripe height
        y = y - stripe_ht
        draw_rectangle(x,y, stripe_ht, stripe_wdt, '#FFFFFF')
        y = y - stripe_ht

    # bottom/last red stripe
    draw_rectangle(x,y, stripe_ht, stripe_wdt, '#B22234')
    y = y - stripe_ht
    

## canton
def draw_canton():
    color = "#3C3B6E"
    height = 200
    width = height*1.9
    c_width = width * (2/5)
    c_height = height * (7/13)
    
    t.penup()
    t.goto(-152, 108)
    t.pendown()
    t.color(color)
    t.begin_fill()

    for i in range(2):
        t.forward(c_width)
        t.right(90)
        t.forward(c_height)
        t.right(90)
    t.end_fill()
    
## stars setup
def draw_stars():
    size = 8
    color = "#FFFFFF"
    
    t.color(color)
    t.begin_fill()
    
    for i in range(5):
        t.forward(size)
        t.right(144)
    t.end_fill()
    
## five rows of 6 stars
def six_stars():
    x1 = - 145
    y1 = 95
    # loop for drawing multiple rows of stars
    for row in range(5):
        for star in range(6):
            t.penup()
            t.goto(x1, y1)
            t.pendown()
            draw_stars()
            x1 = x1 + 25
        # resetting x1 to the starting position
        x1 = - 145
        # moving down to the next row
        # setting the distance between the rows
        y1 = y1 - 20

## four rows of 5 stars
def five_stars():
    x2 = - 132
    y2 = 85
    # loop for drawing multiple rows of stars
    for row in range(4):
        for star in range(5):
            t.penup()
            t.goto(x2, y2)
            t.pendown()
            draw_stars()
            x2 = x2 + 25
        # reseting x1 to the starting position
        x2 = - 132
        # moving down to the next row
        # setting the distance between the rows
        y2 = y2 - 20

draw_stripes()
draw_canton()
six_stars()
five_stars()
t.hideturtle
```

```python
########## French Flag ##########
import turtle 
import time 

scr=turtle.getscreen()
scr.title("French Flag")
scr.bgcolor("#FFFDD0")

t=turtle.Turtle()
t.speed(30)
t.penup()

## rectangles
def draw_rectangle(x,y,height,width,color):
    t.goto(x,y)
    t.pendown()
    t.color(color)
    t.begin_fill()
    
    t.forward(width)
    t.right(90)
    t.forward(height)
    t.right(90)
    t.forward(width)
    t.right(90)
    t.forward(height)
    t.right(90)
    
    t.end_fill()
    t.penup()

## stripes
def draw_stripes():
    x = -152
    y = 108
    height = 200
    width = height*1.42

    stripe_ht = height
    stripe_wdt = width/3

    draw_rectangle(x,y, stripe_ht, stripe_wdt, '#0050a4')
        # decrease value of x by stripe width
    x = x + stripe_wdt
    draw_rectangle(x,y, stripe_ht, stripe_wdt, '#ffffff')
    x = x + stripe_wdt
    draw_rectangle(x,y, stripe_ht, stripe_wdt, '#ef4135')
    x = x + stripe_wdt

draw_stripes()
t.hideturtle

```
